2. Crie uma classe **Veiculo** com os atributos **marca**, **modelo** e **ano**. Crie classes filhas **Carro** e **Moto** que adicionam o atributo **quantidade_de_portas** e **cilindradas**, respectivamente. Crie um método para imprimir as informações do **Veiculo** e um método para imprimir as informações do **Carro** e da **Moto**.

```python
class Veiculo():
  # Propriedades

  # Métodos
  def __init__(self, marca, modelo, ano):
    self._marca = marca
    self._modelo = modelo
    self._ano = ano
  
  def getInformacoes(self):
    print(f"\nMarca: {self._marca}\nModelo: {self._modelo}\nAno: {self._ano}")
  
class Carro(Veiculo):
  # Propriedades

  # Métodos
  def __init__(self, marca, modelo, ano, quantidadePortas):
    super().__init__(marca, modelo, ano)
    self.__quantidadePortas = quantidadePortas

  def getInformacoes(self):
    super().getInformacoes()
    print(f"Tipo: Carro\nQuantidade de Portas: {self.__quantidadePortas}")

class Moto(Veiculo):
  # Propriedades

  # Métodos
  def __init__(self, marca, modelo, ano, cilindradas):
    super().__init__(marca,modelo,ano)
    self.__cilindradas = cilindradas

  def getInformacoes(self):
    super().getInformacoes()
    print(f"Tipo: Moto\nQuantidade de Portas: {self.__cilindradas}")
  

# Objetos
veiculo1 = Veiculo("DAF", "410 Bitruck", "2020")
veiculo1.getInformacoes()

carro1 = Carro("Mitsubishi", "Lancer Evo", "2021", 4)
carro1.getInformacoes()

moto1 = Moto("Yamaha", "150", "2023", "149 cc")
moto1.getInformacoes()  
```

7. Implemente o Diagrama de Classes de Empregados ilustrado abaixo.

Observações:
* O aumento do empregado é o salário multiplicado pelo percentual
* O aumento do chefe é o salário multiplicado pelo percentual + benefícios
* O aumento do estagiário é o salário multiplicado pelo percentual - descontos

image.png


```python
class Empregado():
  # Propriedades

  # Métodos
  def __init__(self, codigo, nome, email, salario):
    self.__codigo = codigo
    self.__nome = nome
    self.__email = email
    self.__salario = salario
  
  def getSalario(self):
    return self.__salario
  
  def setSalario(self, novoSalario):
    self.__salario = novoSalario
  
  def setAumentoSalarial(self, percentual):
    self.__salario += self.__salario * (percentual/100)
    return self.__salario

class Chefe(Empregado):
  # Propriedades

  # Métodos
  def __init__(self, codigo, nome, email, salario, beneficio):
    super().__init__(codigo, nome, email, salario)
    self.__beneficio = beneficio

  def setAumentoSalarial(self, percentual):
    salario = super().setAumentoSalarial(percentual)
    super().setSalario(salario + self.__beneficio) 

class Estagiario(Empregado):
  # Propriedades

  # Métodos
  def __init__(self, codigo, nome, email, salario, descontos):
    super().__init__(codigo, nome, email, salario)
    self.__descontos = descontos
  
  def setAumentoSalarial(self, percentual):
    salario = super().setAumentoSalarial(percentual)
    super().setSalario(salario - self.__descontos) 

# Objetos
empregado1 = Estagiario(794 ,"Roberto", "db_fan_db@gmail.com", 2500, 200)
empregado1.setAumentoSalarial(5)
print(empregado1.getSalario())

empregado2 = Empregado(315, "Lucas", "adrianoielusc@gmail.com", 4000)
empregado2.setAumentoSalarial(10)
print(empregado2.getSalario())

chefe = Chefe(1, "Alfredo", "likaboss@gmail.com", 25000, 3000)
chefe.setAumentoSalarial(40)
print(chefe.getSalario())
```

8. Implemente o código para as classes abaixo:

**a**) crie uma classe chamada **Ingresso** que possui um **valor** em reais e um método
**imprimeValor()**

**b**) crie uma classe **VIP** que herda de **Ingresso** e possui um **valor** adicional. Crie também um método que retorne o **valor** do ingresso **VIP** (como o adicional incluído).

**c**) crie uma classe **Normal**, que herda **Ingresso** e possui um método que **imprime**: "*Ingresso Normal*".

**d**) crie uma classe **CamaroteInferior** (que possui a localização do ingresso e métodos para acessar e imprimir esta localização) e uma classe **CamaroteSuperior**, que é mais cara (possui valor adicional). Esta última possui um método para retornar o valor do ingresso. Ambas as classes herdam a classe **VIP**.

```python
class Ingresso():
  # Propriedades

  # Métodos
  def __init__(self,valor):
    self._valor = valor
  
  def getImprimeValor(self):
    print(f"O valor do ingresso é {self._valor}")
  
class VIP(Ingresso):
  # Propriedades
  _adicional = 30

  # Métodos
  def __init__(self,valor):
    super().__init__(valor + self._adicional)

  def getImprimeValor(self):
    print(f"Ingresso VIP")
    super().getImprimeValor()

class Normal(Ingresso):
  # Propriedades
  
  # Métodos
  def __init__(self,valor):
    super().__init__(valor)

  def getImprimeValor(self):
    print(f"Ingresso Normal")
    super().getImprimeValor()
    print("---"*4)

class CamaroteInferior(VIP):
  # Propriedades
  _local = "Camarote Inferior"

  # Métodos
  def __init__(self,valor):
    super().__init__(valor)

  def getImprimeLocal(self):
    print(f"Ingresso localizado no {self._local}")
    print("---"*4)
  
class CamaroteSuperior(VIP):
  # Propriedades
  _local = "Camarote Superior"
  _adicional = 30
  # Métodos
  def __init__(self,valor):
    super().__init__(valor + self._adicional)

  def getImprimeValor(self):
    super().getImprimeValor()

  def getImprimeLocal(self):
    print(f"Ingresso localizado no {self._local}")
    print("---"*4)

# Objetos

proletariado1 = Normal(25)
proletariado1.getImprimeValor()

burgues1 = CamaroteInferior(25)
burgues1.getImprimeValor()
burgues1.getImprimeLocal()  

burgues2 = CamaroteSuperior(25)
burgues2.getImprimeValor()
burgues2.getImprimeLocal() 
```

**b**) crie a classe **Assistente**, que também é **Funcionário**, e que possui um **número de
matrícula** (*use o método get*).

**c**) sabendo que os Assistentes Técnicos possuem um bônus salarial e que os Assistentes Administrativos possuem um turno (dia ou noite) e um adicional noturno, crie as classes **Tecnico** e **Administrativo**. Para cada um destas classes, imprima o número de matrícula e o nome de cada um deles.

```python
class Funcionario():
  # Propriedades

  # Métodos
  def __init__(self, nome, endereco, telefone, email):
    self.__nome = nome
    self.__endereco = endereco
    self.__telefone = telefone
    self.__email = email

  def getExibeDados(self):
    print(f"\nNome: {self.__nome}\nEndereço: {self.__endereco}\nTelefone: {self.__telefone}\nEmail: {self.__email}")
  
  def getNome(self):
    return self.__nome
class Assistente(Funcionario):
  # Propriedades

  # Métodos
  def __init__(self, nome, endereco, telefone, email, numeroMatricula):
    super().__init__(nome, endereco, telefone, email)
    self.__numeroMatricula = numeroMatricula
  
  def getNumeroMatricula(self):
    return self.__numeroMatricula

class Tecnico(Assistente):
  # Propriedades
  __cargo = "Assistente Técnico"
  # Métodos
  def __init__(self, nome, endereco, telefone, email, numeroMatricula):
    super().__init__(nome, endereco, telefone, email, numeroMatricula)
    
  def getImprimeDados(self):
    super().getExibeDados()
    print(f"Cargo: {self.__cargo}")

class Administrativo(Assistente):
  # Propriedades
  __cargo = "Assistente Administrativo"
  # Métodos
  def __init__(self, nome, endereco, telefone, email, numeroMatricula):
    super().__init__(nome, endereco, telefone, email, numeroMatricula)
    
  def getImprimeDados(self):
    super().getExibeDados()
    print(f"Cargo: {self.__cargo}")

# Objetos
assistente1 = Tecnico("Roberto Carlos", "Rua São Paulo, 763", "(41) 9433 2057", "robertocrl@gmail.com", "09257")
assistente1.getImprimeDados()

assistente2 = Administrativo("Luiz Fontes", "Rua João Luiz, 235", "(44) 9643 4807", "luizfontes@gmail.com", "47856")
assistente2.getImprimeDados()
```

```python
class Pessoa():
  # Propriedades

  # Métodos
  def __init__(self, nome, idade):
    self.__nome = nome
    self.__idade = idade

  def getNome(self):
    return self.__nome
  
  def getIdade(self):
    return self.__idade

class Rica(Pessoa):
  # Propriedades

  # Métodos
  def __init__(self, nome, idade, dinheiro):
    super().__init__(nome, idade)
    self.__dinheiro = dinheiro
  
  def setFazCompras(self,ValorGasto):
    self.__dinheiro -= ValorGasto
    print(f"{super().getNome()} de {super().getIdade()} anos, ainda tem R${self.__dinheiro} reais no banco")
  
class Pobre(Pessoa):
  # Propriedades

  # Métodos
  def __init__(self, nome, idade):
    super().__init__(nome,idade)
  
  def getTrabalha(self):
    print(f"{super().getNome()} aos seus {super().getIdade()} anos, está cansado/a de tanto trabalhar.")

class Miseravel(Pessoa):
  # Propriedades

  # Métodos
  def __init__(self, nome, idade):
    super().__init__(nome,idade)
  
  def getMendiga(self):
    print(f"{super().getNome()} aos seus {super().getIdade()} anos, está mendigando um real.")

# Objetos
rico1 = Rica("Antonio Lopes","58",940000)
rico1.setFazCompras(50000)

pobre1 = Pobre("Mateus Augusto da Silva", "24")
pobre1.getTrabalha()

miseravel1 = Miseravel("Clóvis de Barros Filho", "54")
miseravel1.getMendiga()
```
